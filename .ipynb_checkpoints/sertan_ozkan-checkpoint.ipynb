```python
import os
import imghdr
import shutil

def delete_irrelevant_files(directory, classes):
    for cls in classes:
        class_dir = os.path.join(directory, cls)
        with os.scandir(class_dir) as entries:
            for entry in entries:
                # Delete directories
                if entry.is_dir():
                    print(f'Deleting directory: {entry.path}')
                    shutil.rmtree(entry.path)

                # Check if the file is empty and delete it
                elif os.path.getsize(entry.path) == 0:
                    print(f'Deleting empty file: {entry.path}')
                    os.remove(entry.path)

                # Check if the file is not an image and delete it
                elif not imghdr.what(entry.path):
                    print(f'Deleting non-image file: {entry.path}')
                    os.remove(entry.path)

print("--------Train--------")
delete_irrelevant_files(train_dir, classes)

print("\n--------Test--------")
delete_irrelevant_files(test_dir, classes)
```

```python
class WeatherDataset(Dataset):
    def __init__(self, directory, transform=None):
        self.directory = directory
        self.transform = transform
        self.classes = os.listdir(directory)
        self.class_to_idx = {cls: i for i, cls in enumerate(self.classes)}
        self.images = []
        self.labels = []
        self.load_images()
        
    def load_images(self):
        for cls in self.classes:
            class_dir = os.path.join(self.directory, cls)
            for image_name in os.listdir(class_dir):
                image_path = os.path.join(class_dir, image_name)
                image = Image.open(image_path)
                if self.transform:
                    image = self.transform(image)
                self.images.append(image)
                self.labels.append(self.class_to_idx[cls])
                
    def __getitem__(self, index):
        return self.images[index], self.labels[index]
    
    def __len__(self):
        return len(self.images)
```

```python
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.Grayscale(num_output_channels=3),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = WeatherDataset(train_dir, transform=transform)
test_dataset = WeatherDataset(test_dir, transform=transform)

print(f'Train dataset size: {len(train_dataset)}')
print(f'Test dataset size: {len(test_dataset)}')
```

# MultiBlockCNN model

This code defines a PyTorch deep learning model called MultiBlockCNN for classifying images into different weather categories. The model architecture is based on a pre-trained MobileNetV2 model, with additional layers added on top.

The MultiBlockCNN class takes a single argument, num_classes, which represents the number of classes that the model will classify the images into (in this case, 4: rain, cloudy, sunrise, and shine).

The constructor of the MultiBlockCNN class loads a pre-trained MobileNetV2 model using the torchvision.models module, and freezes the first three blocks of the model to prevent them from being updated during training. Then, it defines a series of convolutional layers and fully connected layers to be added on top of the pre-trained model.

The forward() method of the MultiBlockCNN class takes an input tensor x (which represents an image), and passes it through the series of convolutional and fully connected layers defined in the constructor. The final output is a tensor representing the predicted probabilities for each of the num_classes weather categories.

(Summarized by ChatGPT)

```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
import torchvision.models as models

class MultiBlockCNN(nn.Module):
    def __init__(self, num_classes):
        super(MultiBlockCNN, self).__init__()

        # Load pre-trained MobileNetV2 model
        mobilenet_v2 = models.mobilenet_v2(pretrained=True)
        mobilenet_features = mobilenet_v2.features

        # Freeze the first three blocks of the MobileNetV2 model
        for param in mobilenet_features[:18].parameters():
            param.requires_grad = False

        self.conv1 = nn.Sequential(
            # remaining layers of the mobilenetV2
            mobilenet_features[0],
            mobilenet_features[1],
            mobilenet_features[2],
            # Biases are freezed just like in the François Chollet's Book!
            nn.Conv2d(24, 32, kernel_size=3, padding=1, bias=False),
            nn.ReLU(inplace=True),
            # Batch implemented to the false bias!
            nn.BatchNorm2d(32),
            # You can see the same thing in the rest of the architecture
            nn.Conv2d(32, 128, kernel_size=3, padding=1, bias=False),
            nn.ReLU(inplace=True),
            nn.BatchNorm2d(128),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout(0.5)
        )

        self.conv2 = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, padding=1, bias=False),
            nn.ReLU(inplace=True),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 64, kernel_size=3, padding=1, bias=False),
            nn.ReLU(inplace=True),
            nn.BatchNorm2d(64),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout(0.5)
        )

        self.fc1 = nn.Sequential(
            nn.Linear(64 * 4 * 4, 256),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(256),
            nn.Dropout(0.25)
        )

        self.fc2 = nn.Sequential(
            nn.Linear(256, 512),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(512),
            nn.Dropout(0.125),
        )

        self.fc3 = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
num_classes = 7
model = MultiBlockCNN(num_classes=num_classes).to(device)
```

```python
with open('model.txt', 'w') as f:
    f.write(str(model))
```

```python
from torch.cuda.amp import GradScaler, autocast
scaler = GradScaler()
```

```python
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    running_corrects = 0

    for data, targets in dataloader:
        data = data.to(device)
        targets = targets.to(device)

        # Use autocast for mixed precision
        with autocast():
            outputs = model(data)
            loss = criterion(outputs, targets)
            _, preds = torch.max(outputs, 1)

        # Scale the loss and perform backpropagation
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item() * data.size(0)
        running_corrects += torch.sum(preds == targets.data)

    epoch_loss = running_loss / len(dataloader.dataset)
    epoch_acc = running_corrects.double() / len(dataloader.dataset)

    return epoch_loss, epoch_acc
```

```python
def validate_epoch(model, dataloader, criterion, device):
    model.eval()
    running_loss = 0.0
    running_corrects = 0

    with torch.no_grad():
        for data, targets in dataloader:
            data, targets = data.to(device), targets.to(device)

            # Use autocast for mixed precision
            with autocast():
                outputs = model(data)
                loss = criterion(outputs, targets)
                _, preds = torch.max(outputs, 1)

            running_loss += loss.item() * data.size(0)
            running_corrects += torch.sum(preds == targets.data)

    epoch_loss = running_loss / len(dataloader.dataset)
    epoch_acc = running_corrects.double() / len(dataloader.dataset)

    return epoch_loss, epoch_acc
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=16)
test_dataloader = DataLoader(test_dataset, batch_size=128, shuffle=False, num_workers=16)
```

```python
for i, data in enumerate(train_dataloader):
    print(i, data[0].shape, data[1].shape)
    break

# show shape of the val_dataloader
for i, data in enumerate(test_dataloader):
    print(i, data[0].shape, data[1].shape)
    break
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=0.0001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5, threshold=0.0001)
```

These three lines of code define the loss function, optimizer, and learning rate scheduler that will be used to train the deep learning model.

**criterion = nn.CrossEntropyLoss()**: This line defines the loss function to be used in training the model. In this case, the loss function is Cross Entropy Loss, which is commonly used in multi-class classification problems. The Cross Entropy Loss measures the difference between the predicted probabilities of the classes and the true labels, and tries to minimize this difference during training.

**optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=0.0001)**: This line defines the optimizer used to update the model weights during training. In this case, the optimizer is Adam, which is a commonly used optimizer in deep learning. The optimizer takes as input the model parameters (model.parameters()), the learning rate (lr=0.001), and a weight decay parameter (weight_decay=0.0001) that helps prevent overfitting by penalizing large weights.

**scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5, threshold=0.0001)**: This line defines the learning rate scheduler that will be used during training. The learning rate scheduler reduces the learning rate when the validation loss plateaus, in order to help the model converge to the optimal solution. In this case, the scheduler is ReduceLROnPlateau, which reduces the learning rate by a factor of 0.1 when the validation loss plateaus for 5 epochs (patience=5), and the threshold for measuring the change in the loss is set to 0.0001 (threshold=0.0001). The mode parameter is set to 'min' which means it will monitor the minimum validation loss.

(ChatGPT)

```python
# create empty lists to store loss and accuracy for each epoch
train_loss_list = []
train_acc_list = []
val_loss_list = []
val_acc_list = []

num_epochs = 100
for epoch in range(num_epochs):
    
    train_loss, train_acc = train_epoch(model, train_dataloader, criterion, optimizer, device)
    val_loss, val_acc = validate_epoch(model, test_dataloader, criterion, device)
    scheduler.step(val_loss)
    
    # append the loss and accuracy to the lists
    train_loss_list.append(train_loss)
    train_acc_list.append(train_acc)
    val_loss_list.append(val_loss)
    val_acc_list.append(val_acc)


    print(f"Epoch: {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train_acc_list = [float(t) for t in train_acc_list]
val_acc_list = [float(t) for t in val_acc_list]

train_acc_df = pd.DataFrame({'Epoch': range(len(train_acc_list)), 'Accuracy': train_acc_list, 'Type': 'Train'})
val_acc_df = pd.DataFrame({'Epoch': range(len(val_acc_list)), 'Accuracy': val_acc_list, 'Type': 'Validation'})

acc_df = pd.concat([train_acc_df, val_acc_df])

sns.set_style('darkgrid')

plt.figure(figsize=(10, 6))
sns.lineplot(data=acc_df, x='Epoch', y='Accuracy', hue='Type')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Save the graph to a file
plt.savefig('accuracy_graph.png', dpi=300)

plt.show()
```
